```python
VariablesByTari = namedtuple('VariablesByTari', ['trcal', 'rtcal',
                             'blf', 't1_and_t2', 't1_and_t3'])
Bitrate = namedtuple('Bitrate', ['reader_bitrate', 'tag_bitrate'])
Preamble = namedtuple('Preamble', ['t_sync_preamble', 't_full_preamble', 'tag_preamble_len'])
```

```python
def get_preamble_duration(m, tari, trext):
    variables_by_tari = variables.get_variables_from_tari(tari)
    bitrate = variables.get_bitrate(variables_by_tari.rtcal,
                                    variables_by_tari.blf,
                                    m)
    preamble = variables.get_preamble(tari,
                                      variables_by_tari.rtcal,
                                      variables_by_tari.trcal,
                                      trext, m)
    return preamble.tag_preamble_len / bitrate.tag_bitrate

def get_blf(tari):
    variables_by_tari = variables.get_variables_from_tari(tari)
    return variables_by_tari.blf
```
